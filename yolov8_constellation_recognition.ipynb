```python
# YOLOv8 Constellation Detection Training
# This notebook handles loading your dataset from Google Drive, training YOLOv8, and evaluating the model

import os
import zipfile
from google.colab import drive
import yaml
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow

# Mount Google Drive
drive.mount('/content/drive')

# Define paths
# Update these paths based on your Google Drive structure
DRIVE_PATH = '/content/drive/MyDrive'  # Base drive path
ZIP_PATH = f'{DRIVE_PATH}/constellation_dataset.zip'  # Path to your zipped dataset
WORKSPACE_PATH = '/content/constellation_project'  # Local workspace in Colab

# Create workspace directory
!mkdir -p {WORKSPACE_PATH}

# Extract dataset from Google Drive to workspace
print("Extracting dataset from Google Drive...")
with zipfile.ZipFile(ZIP_PATH, 'r') as zip_ref:
    zip_ref.extractall(WORKSPACE_PATH)

# Dataset directory structure after extraction
DATASET_PATH = f'{WORKSPACE_PATH}/constellation_dataset'  # Based on the provided info

# Verify the dataset structure
print("Checking dataset structure...")
!ls -la {DATASET_PATH}
!ls -la {DATASET_PATH}/train/images
!ls -la {DATASET_PATH}/valid/images
!ls -la {DATASET_PATH}/test/images

# Use the existing data.yaml from the dataset
yaml_path = f'{DATASET_PATH}/data.yaml'

if not os.path.exists(yaml_path):
    raise FileNotFoundError(f"data.yaml not found at {yaml_path}. Please check the path.")

# Load the existing data.yaml
print(f"Loading existing data.yaml from {yaml_path}")
with open(yaml_path, 'r') as f:
    data_yaml = yaml.safe_load(f)

# Update the path to match our current setup
original_path = data_yaml.get('path', '')
data_yaml['path'] = DATASET_PATH

# Make sure train, val and test paths are properly set
if 'train' in data_yaml and data_yaml['train'].startswith('../'):
    data_yaml['train'] = data_yaml['train'].replace('../', '')
if 'val' in data_yaml and data_yaml['val'].startswith('../'):
    data_yaml['val'] = data_yaml['val'].replace('../', '')
if 'test' in data_yaml and data_yaml['test'].startswith('../'):
    data_yaml['test'] = data_yaml['test'].replace('../', '')

# Add 'val' key if missing (using 'valid/images')
if 'val' not in data_yaml or not data_yaml['val']:
    data_yaml['val'] = 'valid/images'

# Print dataset configuration
print(f"Dataset path: {data_yaml['path']}")
print(f"Train path: {data_yaml['train']}")
print(f"Validation path: {data_yaml['val']}")
print(f"Test path: {data_yaml.get('test', 'Not specified')}")
print(f"Number of classes: {len(data_yaml['names'])}")
print(f"Classes: {data_yaml['names']}")

# Save the updated YAML back
with open(yaml_path, 'w') as f:
    yaml.safe_dump(data_yaml, f)

# Print dataset statistics
print("\nDataset Statistics:")
train_imgs = os.listdir(f"{DATASET_PATH}/train/images")
valid_imgs = os.listdir(f"{DATASET_PATH}/valid/images")
test_imgs = os.listdir(f"{DATASET_PATH}/test/images") if os.path.exists(f"{DATASET_PATH}/test/images") else []

print(f"Train set: {len(train_imgs)} images")
print(f"Valid set: {len(valid_imgs)} images")
print(f"Test set: {len(test_imgs)} images")

# Save or update the dataset configuration is no longer needed as we're using the existing data.yaml

# Train YOLOv8 model
# You can customize the model size: 'n' (nano), 's' (small), 'm' (medium), 'l' (large), 'x' (xlarge)
MODEL_SIZE = 'm'  # Medium balance between speed and accuracy
EPOCHS = 100      # Increased for better convergence with 12 zodiac classes
BATCH_SIZE = 16
IMAGE_SIZE = 640  # YOLOv8 default

print(f"Training YOLOv8{MODEL_SIZE} model for {EPOCHS} epochs...")

# Initialize and train the model
model = YOLO(f'yolov8{MODEL_SIZE}.pt')  # Load pretrained model
results = model.train(
    data=yaml_path,
    epochs=EPOCHS,
    batch=BATCH_SIZE,
    imgsz=IMAGE_SIZE,
    patience=10,  # Early stopping patience
    save=True,  # Save results
    project=WORKSPACE_PATH,
    name='constellation_detection',
    exist_ok=True
)

# Evaluate the model on validation data
print("Evaluating model on validation dataset...")
val_results = model.val(data=yaml_path)
print(f"Validation mAP50: {val_results.box.map50}")
print(f"Validation mAP50-95: {val_results.box.map}")

# Test on test dataset
print("Testing model on test dataset...")
test_results = model.predict(
    source=f"{DATASET_PATH}/test/images",
    save=True,
    project=WORKSPACE_PATH,
    name='test_results',
    exist_ok=True
)

# Visualize some test results
test_images = os.listdir(f"{DATASET_PATH}/test/images")
if test_images:
    num_samples = min(5, len(test_images))
    print(f"Displaying {num_samples} test predictions:")

    for i in range(num_samples):
        img_path = f"{DATASET_PATH}/test/images/{test_images[i]}"
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Run prediction on this image
        results = model.predict(source=img_path)

        # Get the result image with boxes
        result_img = results[0].plot()

        # Display
        plt.figure(figsize=(10, 8))
        plt.imshow(result_img)
        plt.title(f"Test image: {test_images[i]}")
        plt.axis('off')
        plt.show()

# Save the trained model
model_path = f"{WORKSPACE_PATH}/constellation_detection/weights/best.pt"
print(f"Best model saved at: {model_path}")

# Copy the model back to Google Drive for persistence
!cp {model_path} {DRIVE_PATH}/best_constellation_model.pt
print(f"Model copied to Google Drive: {DRIVE_PATH}/best_constellation_model.pt")

# Usage example
print("\nExample code to use the trained model:")
print("""
# Load your trained model
from ultralytics import YOLO
model = YOLO('path/to/best_constellation_model.pt')

# Run inference on an image
results = model.predict('path/to/test_image.jpg')

# Display results
import cv2
from google.colab.patches import cv2_imshow
result_img = results[0].plot()
cv2_imshow(result_img)
""")

print("\nTraining and evaluation complete!")
```

```python
def plot_confusion_matrix(model, workspace_path, dataset_path, dataset_type='valid'):
    """
    Function to plot confusion matrix from YOLOv8 results for constellation detection project.

    Parameters:
    - model: Loaded YOLO model
    - workspace_path: Path to the workspace directory ('/content/constellation_project')
    - dataset_path: Path to dataset folder ('/content/constellation_project/constellation_dataset')
    - dataset_type: 'valid' or 'test' or 'train'

    Returns:
    - confusion matrix and class names (displays the confusion matrix)
    """
    import os
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    import yaml
    from sklearn.metrics import confusion_matrix

    # Load class names from data.yaml
    yaml_path = f"{dataset_path}/data.yaml"
    with open(yaml_path, 'r') as f:
        data_config = yaml.safe_load(f)

    class_names = data_config['names']
    num_classes = len(class_names)

    # Set paths
    images_dir = f"{dataset_path}/{dataset_type}/images"
    labels_dir = f"{dataset_path}/{dataset_type}/labels"

    if not os.path.exists(images_dir) or not os.path.exists(labels_dir):
        print(f"Error: {images_dir} or {labels_dir} not found")
        return

    # Collect ground truth and predictions
    y_true = []
    y_pred = []

    images = os.listdir(images_dir)
    for img_file in images:
        img_path = os.path.join(images_dir, img_file)
        # Since all images are jpg, we only need to replace .jpg with .txt
        label_file = os.path.join(labels_dir, img_file.replace('.jpg', '.txt'))

        # Get ground truth
        if os.path.exists(label_file):
            with open(label_file, 'r') as f:
                gt_classes = [int(line.split()[0]) for line in f.readlines()]

            # Skip if no ground truth
            if not gt_classes:
                continue

            # Get prediction
            results = model.predict(img_path, conf=0.25)

            pred_classes = []
            for r in results:
                boxes = r.boxes
                if len(boxes) > 0:
                    confidences = [float(box.conf.item()) for box in boxes]
                    class_ids = [int(box.cls.item()) for box in boxes]

                    if confidences:
                        max_conf_idx = np.argmax(confidences)
                        pred_classes.append(class_ids[max_conf_idx])

            # Add to lists - using first ground truth and first prediction
            if pred_classes:
                y_true.append(gt_classes[0])
                y_pred.append(pred_classes[0])

    # Create confusion matrix
    conf_mat = confusion_matrix(y_true, y_pred, labels=range(num_classes))

    # Plot
    plt.figure(figsize=(12, 10))
    sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names,
                yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Constellation Detection Confusion Matrix')
    plt.tight_layout()
    plt.show()

    # Return for further analysis if needed
    return conf_mat, class_names


# Example usage:

# After training your model:
from ultralytics import YOLO

# Define your paths (matching your project structure)
WORKSPACE_PATH = '/content/constellation_project'
DATASET_PATH = f'{WORKSPACE_PATH}/constellation_dataset'

# Load your model
model = YOLO(f'{WORKSPACE_PATH}/constellation_detection/weights/best.pt')

# Plot confusion matrix
conf_mat, class_names = plot_confusion_matrix(
    model=model,
    workspace_path=WORKSPACE_PATH,
    dataset_path=DATASET_PATH,
    dataset_type='valid'  # or 'test'
)
```

```python
# You can then find specific confusion cases:
# Example: Finding how many times one constellation was confused with another
cancer_idx = class_names.index('Cancer')
virgo_idx = class_names.index('Virgo')
print(f"Cancer classified as Virgo: {conf_mat[cancer_idx, virgo_idx]} times")
```
